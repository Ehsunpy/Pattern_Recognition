```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd


le = LabelEncoder()

df['ethnicity'] = df['ethnicity'].astype(str).str.strip()
df['ethnicity'] = le.fit_transform(df['ethnicity'])

df['contry_of_res'] = df['contry_of_res'].astype(str).str.strip()
df['contry_of_res'] = le.fit_transform(df['contry_of_res'])

df['gender'] = df['gender'].astype(str).str.strip()
df['gender'] = le.fit_transform(df['gender'])


df['relation'] = df['relation'].astype(str).str.strip()
df['relation'] = le.fit_transform(df['relation'])


df['austim'] = df['austim'].astype(str).str.strip()
df['austim'] = le.fit_transform(df['austim'])


df['age_desc'] = df['age_desc'].astype(str).str.strip()
df['age_desc'] = le.fit_transform(df['age_desc'])


df['jundice'] = df['jundice'].astype(str).str.strip()
df['jundice'] = le.fit_transform(df['jundice'])

df['Class/ASD'] = df['Class/ASD'].astype(str).str.strip()
df['Class/ASD'] = le.fit_transform(df['Class/ASD'])


df['used_app_before'] = df['used_app_before'].astype(str).str.strip()
df['used_app_before'] = le.fit_transform(df['used_app_before'])
df.head()
```

```python
df.age.head()
```

```python
df.info()
```

```python
y= df['Class/ASD']
df.drop('Class/ASD', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd


numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df.head()
```

```python
y.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df.corr()
# print(corr_matrix)

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
cov_matrix = df.cov()
print(cov_matrix)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    df, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# دقت روی تست
score = gnb.score(X_test, y_test)
print(f"Test Accuracy: {score:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = gnb.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```
