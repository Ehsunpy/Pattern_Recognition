Using **Min-Max Normalization**, suppose that the minimum and maximum values for the attribute _salary_ are `120,000` Rials and `980,000` Rials, respectively. We would like to map income to the range `0` to `1`.

Calculate the new value of `730,500` Rials using:
1. Min-Max Normalization
2. Z-Score Normalization
3. Normalization by Decimal Scaling

 min-max normalization
$$
x_{normalized} = \frac{x - min}{max - min} \times (new_{max} - new_{min}) + new_{min}
$$

```python
x = 730500
max=980000
min = 12000
```

```python
x_normalize = (x - min) /( max-min)
print(f'Min-Max Normalization = {x_normalize}')
```
